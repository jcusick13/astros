```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import pymc3 as pm
```

```python
df = pd.read_csv('data/astros_bangs_summary.csv')
```

```python
df
```

```python
with pm.Model() as model:
    # Uninformed prior for alpha and beta
    ab = pm.HalfFlat('ab', shape=2, testval=np.asarray([1., 1.]))
    adv_mu = pm.Normal('adv_mu', mu=0, sd=1)
    eps = pm.HalfCauchy('eps', beta=1)


    theta_bang = pm.Beta('theta_bang', alpha=ab[0], beta=ab[1], shape=len(df))
    theta_no_bang = pm.Beta('theta_no_bang', alpha=ab[0], beta=ab[1], shape=len(df))
    
    # OBP for each player
    obp_bang = pm.Binomial(
        'obp_bang', p=theta_bang, observed=df.on_base_bangs, n=df.n_bangs
    )
    obp_no_bang = pm.Binomial(
        'obp_no_bang', p=theta_no_bang, observed=df.on_base_no_bangs, n=df.n_no_bangs
    )
    adv_obs = theta_bang - theta_no_bang
    
    advantage = pm.Normal('advantage', mu=adv_mu, sd=eps, observed=adv_obs)
    trace = pm.sample(5000, tune=1000)
```

```python
pm.traceplot(trace, var_names=['theta_bang', 'theta_no_bang', 'adv_mu'])
```
